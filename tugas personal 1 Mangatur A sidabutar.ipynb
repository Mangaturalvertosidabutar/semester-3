```python
import random
import string

class Password:
    def __init__(self, strength="high", length=None):
        self.strength = strength
        self.length = length if length is not None else self.get_default_length()
        self.password = self.generate_password()

    def get_default_length(self):
        if self.strength == "low":
            return 8
        elif self.strength == "mid":
            return 12
        elif self.strength == "high":
            return 16
        else:
            raise ValueError("Invalid strength specified")

    def generate_password(self):
        if self.strength == "low":
            chars = string.ascii_letters
        elif self.strength == "mid":
            chars = string.ascii_letters + string.digits
        elif self.strength == "high":
            chars = string.ascii_letters + string.digits + string.punctuation
        else:
            raise ValueError("Invalid strength specified")

        return ''.join(random.choice(chars) for _ in range(self.length))

    @classmethod
    def show_input_universe(cls):
        char_pool = {
            "letters": list(string.ascii_letters),
            "numbers": list(string.digits),
            "punctuation": list(string.punctuation)
        }
        return char_pool

# Example Usage
p1 = Password(strength="low")
print(p1.password)

p2 = Password(strength="mid", length=37)
print(p2.password)

p3 = Password(strength="high")
print(p3.password)

p4 = Password()
print(p4.password)
```

1. Berikut adalah beberapa langkah yang dapat membantu Anda dalam proses analisis persyaratan:

1.Identifikasi Stakeholder:
Identifikasi siapa yang terlibat dalam proyek dan siapa yang akan menggunakan perangkat lunak. Ini mungkin termasuk pengguna akhir, manajer, pemilik bisnis, dan pemangku kepentingan lainnya.
2.Kumpulkan Informasi:
Kumpulkan informasi yang relevan dari stakeholder. Ini dapat berupa wawancara, dokumen, catatan rapat, atau survei. Tujuannya adalah untuk memahami tujuan proyek, kebutuhan pengguna, dan masalah yang perlu dipecahkan.
3.Identifikasi Kasus Pengguna:
Identifikasi berbagai kasus pengguna atau skenario yang akan diakomodasi oleh perangkat lunak. Kasus pengguna adalah cara-cara yang berbeda di mana pengguna akan berinteraksi dengan sistem.
4.Identifikasi Entitas dan Hubungan:
Identifikasi entitas (objek) yang akan terlibentuk dalam sistem. Definisikan atribut dan hubungan antara entitas-entitas tersebut. Ini dapat dilakukan dengan menggunakan teknik seperti pemodelan UML (Unified Modeling Language).
5.Analisis Kebutuhan Fungsional:
Identifikasi kebutuhan fungsional yang harus diimplementasikan dalam perangkat lunak. Kebutuhan fungsional adalah deskripsi tentang apa yang sistem harus lakukan. Pisahkan kebutuhan ini berdasarkan objek dan hubungannya.
6.Analisis Kebutuhan Non-Fungsional:
Selain kebutuhan fungsional, identifikasi juga kebutuhan non-fungsional seperti kinerja, keamanan, keandalan, dan sebagainya. Pastikan untuk memahami kriteria yang harus dipenuhi dalam konteks ini.
7.Prioritaskan Persyaratan:
Prioritaskan persyaratan berdasarkan urgensi dan kepentingan. Ini membantu dalam mengarahkan fokus pengembangan dan alokasi sumber daya.
8.Validasi Persyaratan:
Validasi persyaratan dengan stakeholder untuk memastikan bahwa mereka memahami dan setuju dengan deskripsi sistem yang diusulkan. Ini membantu menghindari ketidaksepakatan di masa depan.
9.Dokumentasikan Persyaratan:
Buat dokumen persyaratan yang lengkap dan jelas. Dokumen ini akan menjadi panduan selama fase perancangan dan pengembangan.
10.Perbarui Persyaratan:
Ingatlah bahwa persyaratan dapat berubah seiring waktu. Pastikan untuk mengelola perubahan persyaratan dengan baik dan melibatkan pemangku kepentingan yang relevan.
Proses analisis persyaratan dalam pengembangan berorientasi objek sering melibatkan pemodelan objek dan diagram UML, seperti diagram use case, diagram kelas, dan diagram sekuensi. Hal ini membantu dalam memvisualisasikan dan mengkomunikasikan struktur sistem serta interaksi antar-objek.

2. Cara menggambar diagram Unified Modeling Language (UML) dan contohnya

1.Diagram Use Case:
Diagram use case digunakan untuk menggambarkan interaksi antara sistem dan berbagai aktor (entitas yang berinteraksi dengan sistem).
Contoh: Diagram use case untuk sistem perpustakaan.
2.Diagram Kelas:
Diagram kelas digunakan untuk menggambarkan struktur kelas, atribut, dan hubungan antara kelas dalam sistem.
Contoh: Diagram kelas untuk model perpustakaan.
3.Diagram Sekuensi:
Diagram sekuensi digunakan untuk menggambarkan urutan pesan dan interaksi antar objek dalam skenario tertentu.
Contoh: Diagram sekuensi untuk proses peminjaman buku dalam sistem perpustakaan.
4.Diagram Aktivitas:
Diagram aktivitas digunakan untuk menggambarkan alur kerja atau proses bisnis dalam sistem.
Contoh: Diagram aktivitas untuk proses pendaftaran anggota perpustakaan.
5.Diagram Komponen:
Diagram komponen digunakan untuk menggambarkan komponen fisik atau logis dalam sistem serta hubungan antara komponen tersebut.
Contoh: Diagram komponen untuk aplikasi perpustakaan yang terdiri dari berbagai modul.
6.Diagram Objek:
Diagram objek digunakan untuk menunjukkan instansi dari kelas-kelas tertentu dan hubungan antara objek-objek tersebut.
Contoh: Diagram objek yang menunjukkan buku-buku tertentu dalam perpustakaan pada waktu tertentu.
7.Diagram Statika:
Diagram statika adalah istilah umum yang mencakup diagram kelas, diagram objek, dan diagram komponen, yang semuanya digunakan untuk menggambarkan struktur statis dalam sistem.

3. Cara mengkomunikasikan bagaimana sistem bekerja

1.Dokumentasi Tertulis:
Penjelasan tertulis adalah cara paling dasar dan umum untuk mengkomunikasikan bagaimana sistem bekerja. Ini dapat berupa dokumen kebutuhan, spesifikasi teknis, panduan pengguna, atau laporan desain. Pastikan untuk menyusun dokumen tersebut dengan jelas dan mudah dimengerti.
2.Diagram UML:
Menggunakan diagram UML seperti diagram use case, diagram kelas, diagram sekuensi, atau diagram aktivitas adalah cara yang kuat untuk menggambarkan berbagai aspek sistem, termasuk interaksi, struktur, dan alur kerja.

Membuat prototipe atau mockup dapat membantu dalam mengkomunikasikan bagaimana sistem akan berinteraksi dengan pengguna. Ini membantu pengguna dan pemangku kepentingan untuk memahami antarmuka pengguna, aliran kerja, dan fungsionalitas dasar.
4.Presentasi Visual:
Menggunakan presentasi visual seperti slide PowerPoint atau papan putih untuk memberikan gambaran umum tentang sistem dan bagaimana komponen berinteraksi. Visualisasi sederhana dan gambar dapat membantu audiens memahami konsep dengan cepat.
5.Demonstrasi Langsung:
Melakukan demonstrasi langsung dengan menjalankan perangkat lunak atau sistem yang sedang dalam pengembangan adalah cara yang sangat efektif untuk mengkomunikasikan cara kerja sistem. Ini memungkinkan pengguna dan pemangku kepentingan melihat sistem dalam aksi.

Menggunakan kasus pengguna (use cases) dan cerita pengguna (user stories) adalah cara lain untuk mengkomunikasikan cara sistem akan digunakan dari perspektif pengguna. Ini membantu dalam memahami kebutuhan pengguna dan interaksi dengan sistem.

Mengadakan pertemuan dan lokakarya dengan tim pengembangan, pemangku kepentingan, dan pengguna adalah cara yang baik untuk berdiskusi tentang bagaimana sistem akan berfungsi. Ini memungkinkan untuk pertukaran ide, klarifikasi, dan pemahaman yang lebih baik.
8.Animasi dan Video:
Dalam beberapa kasus, animasi atau video dapat digunakan untuk menggambarkan alur kerja sistem secara visual. Ini berguna terutama ketika sistem memiliki interaksi yang kompleks atau alur kerja yang sulit dipahami dengan mudah.
9.Dokumentasi Interaktif:
Membuat dokumentasi interaktif, seperti situs web atau wiki proyek, dapat memberikan sumber daya yang mudah diakses dan diperbarui yang memungkinkan pemangku kepentingan untuk memahami cara sistem berfungsi.

4. Berikut adalah perbedaan antara keempat konsep tersebut:

1.Kelas (Class):
Kelas adalah cetakan atau blueprint untuk menciptakan objek. Ini mendefinisikan struktur dan perilaku objek yang akan dihasilkan dari kelas tersebut.
Kelas berisi atribut (variabel) dan metode (fungsi) yang menjelaskan karakteristik dan perilaku yang dimiliki oleh objek yang akan dibuat berdasarkan kelas ini.
Contoh: Kelas "Mobil" dapat memiliki atribut seperti "merek," "model," dan metode seperti "menghidupkanMesin" dan "berhenti."
2.Objek (Object):
Objek adalah instansi konkret dari suatu kelas. Ketika Anda membuat objek, Anda sebenarnya membuat instance dari kelas tersebut dengan karakteristik dan perilaku tertentu.
Objek memiliki atribut yang nilainya dapat berbeda-beda antar objek yang berbeda, tetapi mengikuti struktur yang didefinisikan oleh kelas.
Contoh: Jika "Mobil" adalah kelas, maka "Mobil A" dan "Mobil B" adalah dua objek yang mungkin memiliki merek, model, dan perilaku yang berbeda.
3.Atribut (Attribute):
Atribut adalah variabel atau data yang dimiliki oleh sebuah objek atau kelas. Mereka digunakan untuk mendefinisikan karakteristik atau properti dari objek.
Atribut dalam sebuah kelas mendefinisikan data yang akan disimpan dalam objek ketika objek dibuat.
Contoh: Dalam kelas "Mobil," atribut bisa berupa "merek," "model," "warna," dan "kecepatan."
4.Perilaku (Behavior):
Perilaku merujuk pada metode atau fungsi yang dapat dipanggil oleh objek atau kelas. Ini menggambarkan tindakan atau operasi yang dapat dilakukan oleh objek atau kelas.
Perilaku menggambarkan apa yang objek atau kelas dapat lakukan dan bagaimana mereka merespons permintaan.
Contoh: Dalam kelas "Mobil," perilaku bisa termasuk metode "menghidupkanMesin," "berhenti," "gantiGigi," dan "mengemudi."
Dalam OOP, kelas berfungsi sebagai templat yang digunakan untuk membuat objek dengan atribut dan perilaku yang sesuai. Objek adalah instansi konkret dari kelas dan memiliki nilai atribut tertentu serta dapat melakukan perilaku yang didefinisikan dalam kelas tersebut. Atribut adalah data yang mendefinisikan karakteristik objek, sedangkan perilaku adalah metode atau fungsi yang mendefinisikan tindakan yang objek dapat lakukan.

5. Jelaskan beberapa teknik desain OO lebih sering digunakan daripada yang lain. Dalam
   kasus contoh studi, kami fokus pada beberapa:
   a. Mengenkapsulasi fitur ke dalam kelas
   b. Warisan untuk memperluas kelas dengan fitur baru
   c. Komposisi untuk membangun kelas dari objek komponen
   
1.Mengenkapsulasi Fitur ke dalam Kelas (Encapsulation):
Mengenkapsulasi fitur berarti menyembunyikan detail implementasi internal dari objek dan memberikan antarmuka yang terdefinisi dengan baik untuk berinteraksi dengan objek tersebut.
Contoh: Sebuah kelas "Lampu" mungkin memiliki atribut "status" yang hanya dapat diubah melalui metode "nyalakan" dan "matikan." Pengguna objek "Lampu" tidak perlu tahu detail cara lampu bekerja di dalamnya.
2.Warisan untuk Memperluas Kelas dengan Fitur Baru (Inheritance):
Warisan memungkinkan kelas baru (kelas turunan) untuk mewarisi atribut dan metode dari kelas yang sudah ada (kelas induk).
Contoh: Anda memiliki kelas "Kendaraan" yang memiliki atribut "kecepatan" dan metode "bergerak." Anda dapat membuat kelas turunan seperti "Mobil" yang mewarisi atribut dan metode dari "Kendaraan" dan menambahkan atribut tambahan seperti "merek" dan metode tambahan seperti "menghidupkanMesin."
3.Komposisi untuk Membangun Kelas dari Objek Komponen (Composition):
Komposisi melibatkan penggunaan objek dari satu kelas sebagai komponen dalam kelas lain untuk membangun struktur yang lebih kompleks.
Contoh: Dalam sistem perpustakaan, Anda dapat memiliki kelas "Perpustakaan" yang memiliki objek "Rak" sebagai komponen. Setiap "Rak" kemudian berisi objek "Buku." Dengan cara ini, Anda menggunakan komposisi untuk membangun struktur yang lebih besar dari objek-objek yang lebih kecil.
4.Polimorfisme untuk Menggunakan Berbagai Implementasi (Polymorphism):
Polimorfisme memungkinkan Anda untuk memperlakukan objek dari kelas yang berbeda dengan cara yang seragam. Ini berarti Anda dapat menggunakan antarmuka umum untuk berinteraksi dengan objek yang berbeda.
Contoh: Anda memiliki kelas "Hewan" dengan metode "suara." Kelas "Anjing" dan "Kucing" adalah kelas turunan dari "Hewan." Anda dapat menggunakan polimorfisme untuk memanggil metode "suara" pada objek "Anjing" dan objek "Kucing" dengan cara yang serupa.
5.Desain Berbasis Interface (Interface-Based Design):
Desain berbasis interface melibatkan penggunaan antarmuka (interface) untuk mendefinisikan kontrak yang harus dipenuhi oleh kelas yang mengimplementasikannya. Ini memisahkan kontrak dari implementasi.
Contoh: Anda dapat memiliki antarmuka "PengirimPesan" yang mendefinisikan metode "kirimPesan." Kelas "Email" dan "SMS" harus mengimplementasikan antarmuka ini untuk memberikan cara yang berbeda untuk mengirim pesan.
