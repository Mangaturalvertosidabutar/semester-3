```python
import math

class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __repr__(self):
        return f"vector({self.x}, {self.y}, {self.z})"

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y, self.z + other.z)
        else:
            raise TypeError("Unsupported operand type for +")

    def __mul__(self, scalar):
        if isinstance(scalar, (int, float)):
            return Vector(self.x * scalar, self.y * scalar, self.z * scalar)
        else:
            raise TypeError("Unsupported operand type for *")

    __rmul__ = __mul__  # For scalar multiplication in the reverse order (e.g., 2 * vector)

    def __eq__(self, other):
        if isinstance(other, Vector):
            return (self.x, self.y, self.z) == (other.x, other.y, other.z)
        else:
            return False

    def __ne__(self, other):
        return not self == other

    def __lt__(self, other):
        if isinstance(other, Vector):
            return self.magnitude() < other.magnitude()
        else:
            raise TypeError("Unsupported operand type for <")

    def __le__(self, other):
        if isinstance(other, Vector):
            return self.magnitude() <= other.magnitude()
        else:
            raise TypeError("Unsupported operand type for <=")

    def __gt__(self, other):
        if isinstance(other, Vector):
            return self.magnitude() > other.magnitude()
        else:
            raise TypeError("Unsupported operand type for >")

    def __ge__(self, other):
        if isinstance(other, Vector):
            return self.magnitude() >= other.magnitude()
        else:
            raise TypeError("Unsupported operand type for >=")

    def __bool__(self):
        return bool(self.magnitude())

    def __getitem__(self, key):
        if key == 'x':
            return self.x
        elif key == 'y':
            return self.y
        elif key == 'z':
            return self.z
        else:
            raise KeyError("Invalid key")

    def __hash__(self):
        return hash((self.x, self.y, self.z))

# Example usage:
v1 = Vector(1, 2, 3)
v2 = Vector(4, 5, 6)
print(v1 + v2)        # vector(5, 7, 9)
print(v1 * 2)         # vector(2, 4, 6)
print(2 * v1)         # vector(2, 4, 6)
print(v1 == v2)       # False
print(v1 < v2)        # True (based on magnitude)
print(v1['y'])        # 2
```

1. Basic Inheritance (Pewarisan Dasar):
Basic Inheritance adalah konsep dalam pemrograman berorientasi objek di mana sebuah kelas baru (kelas turunan atau subclass) dapat mewarisi atribut dan metode dari kelas yang sudah ada (kelas dasar atau superclass). Dengan pewarisan, kelas turunan dapat memanfaatkan fungsionalitas yang sudah ada dalam kelas dasar, yang memungkinkan untuk mengorganisasi kode dengan lebih efisien dan memungkinkan penggunaan kembali kode.

2. Multiple Inheritance (Pewarisan Ganda):
Multiple Inheritance adalah konsep di mana sebuah kelas dapat mewarisi atribut dan metode dari lebih dari satu kelas dasar. Dalam multiple inheritance, sebuah kelas turunan memiliki dua atau lebih kelas dasar. Ini memungkinkan kelas turunan untuk menggabungkan fungsionalitas dari beberapa sumber yang berbeda. Namun, multiple inheritance bisa menjadi kompleks karena dapat muncul konflik jika kelas dasar memiliki metode atau atribut dengan nama yang sama.

Dalam kedua jenis pewarisan, tujuannya adalah untuk memungkinkan penggunaan kembali kode, mengorganisasi kode dengan lebih baik, dan menciptakan hierarki kelas yang lebih kompleks. Pemahaman yang baik tentang pewarisan adalah konsep penting dalam pemrograman berorientasi objek.

1. Inheritance from Built-In (Pewarisan dari Built-In):
Ini adalah kemampuan dalam pemrograman berorientasi objek untuk mewarisi sifat dan perilaku dari kelas-kelas bawaan atau built-in yang telah ada dalam bahasa pemrograman. Contohnya, dalam Python, kelas-kelas seperti list, dict, atau str dapat diwarisi atau diperluas untuk membuat kelas-kelas turunan khusus dengan fitur tambahan.

2. Polymorphism (Polimorfisme):
Polimorfisme adalah konsep dalam pemrograman berorientasi objek yang memungkinkan objek dari kelas-kelas yang berbeda untuk merespons metode yang sama dengan cara yang sesuai dengan kelas masing-masing. Ini memungkinkan kode untuk menjadi lebih fleksibel dan dapat digunakan untuk bekerja dengan berbagai tipe objek tanpa harus tahu tipe objek secara eksplisit.

3. Duck Typing (Tipe Bebek):
Duck typing adalah filosofi dalam pemrograman yang berfokus pada perilaku objek daripada tipe objek. Konsep ini menyatakan bahwa objek dikenali berdasarkan perilaku yang mereka tunjukkan, bukan berdasarkan tipe atau kelas mereka. Jika suatu objek berjalan seperti itik dan bersuara seperti itik, maka itu dianggap sebagai itik. Dalam banyak bahasa pemrograman yang mendukung duck typing, pengkodean lebih fleksibel dan terfokus pada apa yang objek dapat lakukan daripada tipe mereka.

```python
# challenge 6 
class Tablet:
    def __init__(self, model):
        if model not in ['lite', 'pro', 'max']:
            raise ValueError("Invalid model type")
        
        self.model = model
        self.base_storage = 32 * (2 ** ['lite', 'pro', 'max'].index(model))
        self.memory = 2 * (2 ** ['lite', 'pro', 'max'].index(model))
        self.added_storage = 0

    def add_storage(self, additional_storage):
        total_storage = self.base_storage + self.added_storage
        if total_storage + additional_storage > 1024:
            raise ValueError("Storage capacity exceeded (max 1024GB)")
        self.added_storage += additional_storage

    @property
    def storage(self):
        return self.base_storage + self.added_storage

    @storage.setter
    def storage(self, value):
        if value < self.memory:
            raise ValueError("Memory cannot exceed storage")
        
        if value > 1024:
            raise ValueError("Storage capacity exceeded (max 1024GB)")
        
        self.base_storage = value - self.memory

    def __repr__(self):
        return f"Tablet(model='{self.model}')"


t1 = Tablet('lite')
print(t1.model)          
print(t1.base_storage)   
print(t1.memory)         
print(t1.added_storage)  

t1.add_storage(64)
print(t1.added_storage)  

t1.storage = 128
print(t1.base_storage)   
print(t1.added_storage)   

print(t1) 
```
